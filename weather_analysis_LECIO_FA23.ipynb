```python
from re import X
from sklearn.model_selection import train_test_split

df_filtered = df[(df['7P'] != 0) & (df['14P'] != 0) & (df['28P'] != 0)]

new_df = df_filtered[['AlfaY', 'AlfaD']]

y_7 = df['7P']

X_train_7, X_test_7, y_train_7, y_test_7 = train_test_split(new_df, y_7, test_size=0.2, train_size = 0.8)

y_14 = df['14P']

X_train_14, X_test_14, y_train_14, y_test_14 = train_test_split(new_df, y_14, test_size=0.2, train_size = 0.8)

y_28 = df['28P']

X_train_28, X_test_28, y_train_28, y_test_28 = train_test_split(new_df, y_28, test_size=0.2, train_size = 0.8)

X_train_7
```

```python
#Ada
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import classification_report
clf = AdaBoostClassifier(n_estimators=67, random_state=34)
ada_7 = clf.fit(X_train_7, y_train_7)
print("Report_7", classification_report(y_test_7, ada_7.predict(X_test_7), zero_division=1))
ada_14 = clf.fit(X_train_14, y_train_14)
print("Report_14", classification_report(y_test_14, ada_14.predict(X_test_14), zero_division=1))
ada_28 = clf.fit(X_train_28, y_train_28)
print("Report_28", classification_report(y_test_28, ada_28.predict(X_test_28), zero_division=1))

```

```python
#Tree
from sklearn.tree import DecisionTreeClassifier
clf_7 = DecisionTreeClassifier(random_state=0, max_depth= 60, min_samples_split = 23)
tree_7 = clf_7.fit(X_train_7, y_train_7)
print("Report_7", classification_report(y_test_7, tree_7.predict(X_test_7),zero_division=1 ))
clf_14 = DecisionTreeClassifier(random_state=0, max_depth= 60, min_samples_split = 23)
tree_14 = clf_14.fit(X_train_14, y_train_14)
print("Report_14", classification_report(y_test_14, tree_14.predict(X_test_14), zero_division=1))
clf_28 = DecisionTreeClassifier(random_state=0, max_depth= 60, min_samples_split = 23)
tree_28 = clf_28.fit(X_train_28, y_train_28)
print("Report_28", classification_report(y_test_28, tree_28.predict(X_test_28), zero_division=1))
```

```python
#Gaussian
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.metrics import classification_report
from sklearn.gaussian_process.kernels import RBF
sample_size = 1000  # Adjust the sample size based on your available memory
X_train_7_sample = X_train_7.sample(n=sample_size, random_state=42)
y_train_7_sample = y_train_7.sample(n=sample_size, random_state=42)
X_train_14_sample = X_train_14.sample(n=sample_size, random_state=42)
y_train_14_sample = y_train_14.sample(n=sample_size, random_state=42)
X_train_28_sample = X_train_28.sample(n=sample_size, random_state=42)
y_train_28_sample = y_train_28.sample(n=sample_size, random_state=42)
clf = GaussianProcessClassifier(kernel=1.0 * RBF(1.0),random_state=0)
gau_7 = clf.fit(X_train_7_sample, y_train_7_sample)
print("Report_7", classification_report(y_test_7, gau_7.predict(X_test_7), zero_division=1))
gau_14 = clf.fit(X_train_14_sample, y_train_14_sample)
print("Report_14", classification_report(y_test_14, gau_14.predict(X_test_14), zero_division=1))
gau_28 = clf.fit(X_train_28_sample, y_train_28_sample)
print("Report_28", classification_report(y_test_28, gau_28.predict(X_test_28), zero_division=1))
```

```python
#NeuralNet
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import classification_report

clf = MLPClassifier(random_state=34, max_iter=500000, learning_rate_init = 10)
a = clf.fit(X_train_7, y_train_7)
print("Report_7", classification_report(y_test_7, a.predict(X_test_7), zero_division=1))
b = clf.fit(X_train_14, y_train_14)
print("Report_14", classification_report(y_test_14, b.predict(X_test_14), zero_division=1))
c = clf.fit(X_train_28, y_train_28)
print("Report_28", classification_report(y_test_28, c.predict(X_test_28), zero_division=1))
```

```python
usable_df = []
for i in range(10):
  a = random.choice(df.columns)
  usable_df.append(a)
print(usable_df)
usable_df = ['AlfaX', 'AlfaY', 'AlfaZ', 'AlfaD']
```

```python
column_names = df.columns
usable_dfs = []

for i in range(1, len(column_names) - 1, 4):
    sublist = list(column_names[i:i+4])
    usable_dfs.append(sublist)

print("usable_dfs:", usable_dfs)
```

```python
for i in usable_dfs:
  print(i)
  pairplot = sb.pairplot(df[i])

  # Extract the individual axes
  axes = pairplot.axes

  # Calculate R-squared for each pair and collect the values
  r_squared_values = []

  for i, row in enumerate(axes):
      for j, ax in enumerate(row):
          if i != j:  # Exclude self-comparisons on the diagonal
              x_var = df.columns[i]
              y_var = df.columns[j]
              x_data = df[x_var]
              y_data = df[y_var]

              if pd.api.types.is_numeric_dtype(x_data) and pd.api.types.is_numeric_dtype(y_data):
                  # Calculate the correlation coefficient (Pearson correlation)
                  r = np.corrcoef(x_data, y_data)[0, 1]

              # Calculate R-squared and append to the list
                  r_squared = r**2
                  r_squared_values.append(r_squared)

  # Compute the average R-squared value
  average_r_squared = sum(r_squared_values) / len(r_squared_values)

  print("Phonetic type: ", i, "Average R-squared value:", average_r_squared)
```
